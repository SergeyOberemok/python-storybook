```python
import matplotlib.pyplot as plt
import sympy as sp
from faker import Faker
```

```python
fake = Faker()
```

---

# Simple

```python
def getRandomPoint(min: int, max: int) -> tuple[int, int]:
    return fake.random_int(min, max), fake.random_int(min, max)
```

## One

```python
boundary = 10
x, y = getRandomPoint(1, boundary)

plt.plot(x, y, 'r.')

plt.axis('square')
plt.axis([0, boundary, 0, boundary])
plt.grid()

plt.show()
```

## Multiple

```python
count = 100
boundary = 10

for i in range(count):
    x, y = getRandomPoint(1, boundary)
    plt.plot(x, y, 'o')

plt.axis([0, boundary, 0, boundary])
plt.grid()

plt.show()
```
