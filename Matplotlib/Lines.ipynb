# Line

## from 0

```python
x = 3
y = 5

plt.plot(x, y, 'ro')
plt.plot([0, x], [0, y])

plt.show()
```

## two points

```python
p1 = (-3, -1)
p2 = (4, 4)

xpoints = [p1[0], p2[0]]
ypoints = [p1[1], p2[1]]

plt.plot(xpoints, ypoints, 'red', linewidth = 5)
plt.plot([4, -3], [-1, 4], linewidth = 5) 

plt.show()
```
