# Population

## Data

### Labels

```python
yearsCount = 8
endYear = 2030
interval = 10

years = [*range(endYear - interval * yearsCount, endYear, interval)]

print(years)
```

### Measures

```python
def getSortedNumbers(size, minNumber, maxNumber):
    numbers = [fake.random_int(minNumber, maxNumber) for _ in range(size)]
    numbers.sort()
    return numbers

print(getSortedNumbers(size, 5, 100))
```

```python
countriesCount = 5

populations = reduce(lambda acc, _: acc | {fake.country(): getSortedNumbers(yearsCount, 5, 100)}, range(countriesCount), {}) 

print(populations)
```

## Chart

```python
_, axis = plt.subplots()
axis.stackplot(years, populations.values(), labels=populations.keys())
axis.set_title('Countries populations')

plt.show()
```
