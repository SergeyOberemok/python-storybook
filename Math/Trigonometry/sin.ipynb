```python
import numpy as np
from numpy import pi
import matplotlib.pyplot as plt
```

```python
def getSinArray(start, stop, quantity):
    x = np.linspace(start, stop, quantity)
    f = np.sin(x)
    
    return x, f
```

```python
start = 0
quantity = 100
a, b = getSinArray(start, pi, quantity)

plt.plot(a, b)
plt.axis('square')
plt.axis([0, pi * 2, -2, 2])
plt.grid()

plt.show()
```
