```python
import matplotlib.pyplot as plt
import math
```

## Line

### With two points

```python
p1 = (-3, -1)
p2 = (4, 4)

xpoints = [p1[0], p2[0]]
ypoints = [p1[1], p2[1]]

plt.plot(xpoints, ypoints, 'red', linewidth = 5)
plt.plot([4, -3], [-1, 4], linewidth = 5) 

plt.show()
```

### From zero

```python
x = 3
y = 5

plt.plot(x, y, 'ro')
plt.plot([0, x], [0, y])

plt.show()
```

### Descartes system

```python
size = 5
plt.axis('square')
plt.axis([-size, size, -size, size])
plt.grid()
plt.plot([-size, size], [0, 0], 'black')
plt.plot([0, 0], [-size, size], 'black')

plt.plot([-2, -2, 2, 2, -2], [-2, 2, 2, -2, -2])

plt.show()
```

### sqrt of x

```python
maxNumber = 20

for x in range(-maxNumber, maxNumber + 1):
    # y = math.sqrt(abs(x))
    y = abs(x) ** (1/2)
    plt.plot([0, x], [0, y])

plt.show()
```
