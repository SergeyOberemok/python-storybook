```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = [[1, 2], [3, 4]]

plt.imshow(A)

plt.xticks([0, 1])
plt.yticks([0, 1])

plt.show()
```

### Text, ticks and color maps

```python
A = np.zeros((10, 10))
xLength = np.shape(A)[0]
yLength = np.shape(A)[1]

for i in range(0, xLength):
    for j in range(0, yLength):
        A[i,j] = 3 * i - 4 * j

plt.imshow(A)

plt.xticks(list(range(0, xLength)))
plt.yticks(list(range(0, yLength)))

for i in range(0, xLength):
    for j in range(0, yLength):
        plt.text(j, i, int(A[i, j]), horizontalalignment='center', verticalalignment='center')

plt.set_cmap('Blues')

plt.show()
```

### Excersice

```python
size = 10
C = np.zeros((size, size))
xLength = np.shape(C)[0]
yLength = np.shape(C)[1]

for i in range(0, xLength):
    for j in range(0, yLength):
        C[i, j] = (-1)**(i + j)

plt.imshow(C)

plt.set_cmap('Greys')
plt.tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)

plt.show()
```
