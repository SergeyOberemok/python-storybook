```python
import sympy as sp
```

### Plotting

```python
from sympy.abc import x

y = x**2

plot = sp.plotting.plot(y)
```

### Amend plot

```python
from sympy.abc import x, y

exp = x/y

yLimits = (-5, 5)

plot = sp.plotting.plot(exp.subs(x, 1), (y,) + yLimits, show = False)

plot.ylim = list(yLimits)
plot.show()
```

### Extend plot

```python
plotToExtend = sp.plotting.plot(exp.subs(x, 2), show = False)
plot.extend(plotToExtend)

plot.show()
```

### Excercise

```python
a = sp.symbols('a')

y = a/(x**2 - a)

plot = None
xLimits = (-5, 5)
yLimits = (-10, 10)

for i in range(1, 5):
    iPlot = sp.plotting.plot(y.subs(a, i), (x,) + xLimits, show = False)
    iPlot[0].label = '$%s$' %sp.latex(y.subs(a, i))
    
    if plot is None:
        plot = iPlot
    else:
        plot.extend(iPlot)

plot.ylim = list(yLimits)
plot.legend = True
plot.show()
```
