```python
import numpy as np
import matplotlib.pyplot as plt
```

### Range

```python
x = range(-3, 4)
y = []

for i in range(0, len(x)):
    result = 2 - x[i]**2
    y.append(result)
    
plt.plot(x, y, 's-')
plt.show()
```

### Linspace - discrete numbers

```python
discreteQuantity = 40
x = np.linspace(-3, 4, discreteQuantity)
y = 2 - np.power(x, 2)

plt.plot(x, y, 'o-')
plt.show()
```

### Comparing

```python
x1 = range(-3, 4)
x2 = np.linspace(-3, 4, 40)
y1 = 2 - np.power(x1, 2)
y2 = 2 - np.power(x2, 2)

plt.plot(x2, y2, 'oy-')
plt.plot(x1, y1, 'sb-')
plt.show()
```
